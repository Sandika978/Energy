```python
import pandas as pd
```

```python
path_dataset='/content/Energy production by different source.csv'
```

```python
df=pd.read_csv(path_dataset)
```

```python
df.head
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
plt.plot(df['No. of Month'], df['wind_iowa'],label='wind_iowa')
plt.plot(df['No. of Month'], df['wind_texas'], label='Texsas')
plt.plot(df['No. of Month'], df['wind_west_north_central'], label='WNC')
plt.xlabel('No. of Month')
plt.ylabel('Wind power Generation in MW')
plt.legend()
plt.title('Wind Power Generation in US states')
plt.savefig('Wind.png')
plt.show()
```
