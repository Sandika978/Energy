```python
## Import Pandas file which read and write CSV, Excel etc. file with labeled and
##time series data

import pandas as pd
```

```python
path_dataset='/content/drive/MyDrive/Colab Notebooks/Energy production by different source.csv'
```

```python
df=pd.read_csv(path_dataset)
```

```python
df.head
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
plt.plot(df['No. of Months'], df['wind_iowa'],label='Iowa')
plt.plot(df['No. of Months'], df['wind_texas'], label='Texsas')
plt.plot(df['No. of Months'], df['wind_west_north_central'], label='WNC')
plt.xlabel('No. of Months from Year Jan 2001 to Feb 2023')
plt.ylabel('Wind power Generation in MW')
plt.legend()
plt.title('Wind Power Generation in US states')
plt.savefig('Wind.png')
plt.show()
```

The graph shows the wind power generation by different states of United States which is used for Electricity.
It shows the US is stepping foot year by year toward the net zero emissions. It is the second country in the world after China in wind power generation.
